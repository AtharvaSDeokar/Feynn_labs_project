### Range and Price factor for all available EV models in India

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\\Users\Atharva.Deokar\Downloads\EVIndia.csv')
```

```python
df
```

```python
plt.barh(df.Car, df.PriceRange, color="red", height=0.7)
plt.xlabel("Price Range")
plt.ylabel("EV Name")
plt.title("Price vs EV Name")
plt.show()
```

```python
plt.barh(df.Car, df.Range, color="green")
plt.xlabel("EV Range")
plt.ylabel("EV Name")
plt.title("EV vs Range")
plt.show()
```

### Observations based on EV sales by Region and Range

1. From above dataset it can be seen that Price of EV in India exist in range 12.24 lakh to 1.8 Cr.
   Considering Indian market which is very Price sensitive, for a new EV startup it is advisible that they should consider the      closely price factor as it is one of the deciding factors before buying an EV.
   New EV startups should closely monitor all the resources and do a thorough RnD before selecting parts and manufacturing          process for their 1st EV.
   Manufacturing cost and EV parts play a major role while deciding the total cost of 1 complete EV unit.
   
2. Range is 2nd most important factor that can affect the EV sales. Indian market / mentally is more inclinded towards an EV        that can be both price effective and has a long range.
   One another factor related to EV sales could be charging time taken by an individual unit to get fully powered up.
   Only having a long range cannot be enough but it should also have a adequately less power up time to please the customer.
   Range and charging time both these factors are corelated and go hand in hand as factors that can influence EV sales.

3. According to study done by Economic Times, on an average for consumers in India, a price point of Rs 23 lakh (or USD 31,000),    a charge time of 35 minutes and a range of 401 kilometers from a single charge represent the 'tipping points' to achieve        mainstream EV adoption. Consumers in India are seeking a lower price point for EVs than consumers in other countries, with      the global average tipping price point for EVs being at USD 36,000 (around Rs 27 lakh), it said.
